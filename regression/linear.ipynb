```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 10.0
b = -2.0

def linear_data(x):
    return  m * x + b + np.random.normal(0, 10, len(x)) # Adding noise to make it interesting

# ()
x = np.linspace(-10, 10, 100)
y = linear_data(x)

X = np.vstack([x, np.ones(len(x))]).T

params = np.linalg.inv(X.T @ X) @ X.T @ y

print(f"m: {params[0]}, b: {params[1]}")

pred = x * params[0] + params[1]

plt.plot(x, y, 'o', label='Data')
plt.plot(x, pred, 'r', label='Fitted line')
plt.legend()
plt.show()
```
